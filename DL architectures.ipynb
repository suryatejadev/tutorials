We can improve the performance of deepnets in 2 ways:
- Increase the depth.  
- Increase the width (number of kernel layers).  

Disadvantage?  
- Large number of params.  
- More prone to overfitting.  
- More computationally expensive.  
- Possible waste of parameters (most weights end up zero).  

Solution?  
- Move from fully connected to sparsely connected architectures, **even inside convolutions**.  

Theoritical Insights  
- Work of Aurora et al: If the PDF of the data is represented by a large , very sparse DNN, then the optimal network topology can be constructed layer by layer by analyzing the correlation statistics of the activations pof the last layer and clustering neurons with highly correlated outputs.  
- This is a strict math proof that requires very strong conditions. 
- However, it resonates with the Hebbian principle: Neurons that fire together, wire together, suggesting it's  applicable even under less strict conditions in practice.  

Practical Issue: 
- Numerical calculation of non-uniform sparse data structures is very inefficient.  
- The DL libraries can efficiently compute dense matrix operations using CPU/GPU.  
- Even normal convolutions are implemented as collections of dense connections to patches in the earlier layer. 

Objective: 
- **An architecture that makes use of the extra sparsity, even at filter level, but exploits the current hardware by utilizing computations on dense matrices.**  
- Literature (3): clustering sparse matrices into relatively dense sub matrices tends to give SOTA performance for sparse matrix multiplication.  
- **Inception Architecture**: The main idea is based on finding out how an optimal local sparse structure in a CNN can be approximated and covered by readily available dense components.
